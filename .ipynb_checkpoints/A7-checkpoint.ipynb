## Cleaning the Data 
Now that the data set has been sucessfully imported, after briefly looking through how the content is layed out and presented to us, a couple of unneccessary elements that are not needed in the such as IP Address, name and survey start/end date. 

*ChatGPT was asked to assist in cleaning the data by providing the original dataset and asking to remove the following: Start Date, End Date, Collector ID, IP Address, Email Address, First Name, Last Name, Custom Data 1, Age, and Birth Sex

```python
# List of explicitly named columns to remove
columns_to_remove = [
    "Start Date", "End Date", "IP Address", "Email Address",
    "First Name", "Last Name", "Custom Data 1", "What is your age?", "Birth Sex?"
]

# Remove unwanted "Unnamed" columns related to age and birth sex
unnamed_columns_to_remove = [
    col for col in climbing_data.columns if ("Unnamed" in col and 
                                             ("age" in col.lower() or "birth sex" in col.lower()))
]

# Remove specific "Unnamed" columns (10 to 16)
unnamed_range_to_remove = [f"Unnamed: {i}" for i in range(10, 17)]

# Combine all lists of columns to remove
columns_to_remove.extend(unnamed_columns_to_remove)
columns_to_remove.extend(unnamed_range_to_remove)

# Drop specified columns
cleaned_data = climbing_data.drop(columns=columns_to_remove, errors="ignore")

# Define renaming map for unnamed columns
rename_mapping = {
    **{f"Unnamed: {i}": "How many years have you climbed?" for i in range(18, 23)},			
    **{f"Unnamed: {i}": "What grade do you consider you climb?" for i in range(24, 29)},
    **{f"Unnamed: {i}": "What grade do you consider you boulder?" for i in range(30, 35)},
    **{f"Unnamed: {i}": "Have you had any climbing injuries? (check all that applies)" for i in range(36, 42)},
    **{f"Unnamed: {i}": "How long did the whole recovery process take before you were climbing at your same grade or better?" for i in range(44, 51)}, 
    **{f"Unnamed: {i}": "Did you require surgery? And are you satisfied with the surgery?" for i in range(52, 56)}  # Corrected range syntax
}

# Apply renaming to cleaned_data
cleaned_data.rename(columns=rename_mapping, inplace=True)

# Export to a CSV for inspection
cleaned_data.to_csv('Cleaned_Climbing_Injuries.csv', index=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load your dataset (replace with your actual file path)
df = pd.read_csv('low_experience_injury_data.csv')

# Step 2: Extract the recovery time columns
recovery_time_columns = [
    "How long did the whole recovery process take before you were climbing at your same grade or better?",
    "How long did the whole recovery process take before you were climbing at your same grade or better?.1",
    "How long did the whole recovery process take before you were climbing at your same grade or better?.2",
    "How long did the whole recovery process take before you were climbing at your same grade or better?.3",
    "How long did the whole recovery process take before you were climbing at your same grade or better?.4",
    "How long did the whole recovery process take before you were climbing at your same grade or better?.5",
    "How long did the whole recovery process take before you were climbing at your same grade or better?.6",
    "How long did the whole recovery process take before you were climbing at your same grade or better?.7"
]

# Step 3: Filter the dataset for 'Hand & Elbow' injury type
df_hand_elbow = df[df['Hand & Elbow'].notnull()]

# Step 4: Extract the recovery times for 'Hand & Elbow'
recovery_times_hand_elbow = df_hand_elbow[recovery_time_columns].stack().reset_index(drop=True)

# Step 5: Convert recovery times to numeric, ignoring non-numeric values
recovery_times_hand_elbow = pd.to_numeric(recovery_times_hand_elbow, errors='coerce')

# Step 6: Plot the distribution of recovery times
plt.figure(figsize=(10, 6))
sns.histplot(recovery_times_hand_elbow, bins=20, kde=True, color='skyblue', label='Recovery Times')

# Step 7: Add vertical line for the most common recovery time (mode)
most_common_recovery_time = recovery_times_hand_elbow.mode()[0]
plt.axvline(most_common_recovery_time, color='red', linestyle='--', label=f'Most Common: {most_common_recovery_time} months')

# Step 8: Add labels and title
plt.title("Distribution of Recovery Times for Hand & Elbow Injuries")
plt.xlabel("Recovery Time (Months)")
plt.ylabel("Frequency")
plt.legend()

# Show the plot
plt.show()
```

After the dataset only contains inputs from the higher level of experience, we can then keep count and determine the most common injuries that occur for these climbers.



```python
# Step 1: Adjust year columns renaming
import pandas as pd


high_experience_data = pd.read_csv('fully_cleaned_climbing_data.csv')


# Step 1: Filter rows to retain only those with input in either "4-5, 6-8, 9-15, or 15+" years
filtered_data = high_experience_data.dropna(
    subset=['How many years have you climbed? (4-5)', 'How many years have you climbed? (6-8)',
           'How many years have you climbed? (9-15)', 'How many years have you climbed? (15+)'],
    how='all'
)

# Step 2: Define the injury columns
injury_columns = [
    "Hand & Elbow", "Shoulder", "Spine", 
    "Pelvis/Hip", "Knee", "Foot & Ankle", "Other"
]

# Step 3: Count the total inputs for each injury type
injury_counts = filtered_data[injury_columns].notna().sum().to_dict()

# Step 4: Identify the most common injury type
most_common_injury = max(injury_counts, key=injury_counts.get)
highest_count = injury_counts[most_common_injury]

# Save the filtered data to a new CSV
filtered_data.to_csv('high_experience_common_injury.csv', index=False)
high_experience_data.head()
# Print results
print("Injury counts for higher experience participants:")
print(injury_counts)
print(f"The most common injury is '{most_common_injury}' with {highest_count} occurrences.")
```

This code followed very similarly to the lower levels of experience method. However, the column headers to find the input for the 4-5, 6-8, 9-15, & 15+ years were inputted instead of the 0-1 and 2-3 years of experience.

The code below will attempt to visualize the recovery times for each injury, highlighting the top 3 most common injuries alongside their typical recovery times using a histogram and kernel density to estimate. 
*The code follows a very similar structure to the visualizations that were created for low-experienced climbers. 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load your dataset (replace with your actual file path)
df = pd.read_csv('high_experience_common_injury.csv')

# Step 2: Extract the recovery time columns
recovery_time_columns = [
    "How long did the whole recovery process take before you were climbing at your same grade or better?",
    "How long did the whole recovery process take before you were climbing at your same grade or better?.1",
    "How long did the whole recovery process take before you were climbing at your same grade or better?.2",
    "How long did the whole recovery process take before you were climbing at your same grade or better?.3",
    "How long did the whole recovery process take before you were climbing at your same grade or better?.4",
    "How long did the whole recovery process take before you were climbing at your same grade or better?.5",
    "How long did the whole recovery process take before you were climbing at your same grade or better?.6",
    "How long did the whole recovery process take before you were climbing at your same grade or better?.7"
]

# Step 3: Filter the dataset for 'Shoulder' injury type
df_shoulder = df[df['Hand & Elbow'].notnull()]

# Step 4: Extract the recovery times for 'Shoulder'
recovery_times_shoulder = df_shoulder[recovery_time_columns].stack().reset_index(drop=True)

# Step 5: Convert recovery times to numeric, ignoring non-numeric values
recovery_times_shoulder = pd.to_numeric(recovery_times_shoulder, errors='coerce')

# Step 6: Plot the distribution of recovery times
plt.figure(figsize=(10, 6))
sns.histplot(recovery_times_shoulder, bins=20, kde=True, color='lightgreen', label='Recovery Times')

# Step 7: Add vertical line for the most common recovery time (mode)
most_common_recovery_time = recovery_times_shoulder.mode()[0]
plt.axvline(most_common_recovery_time, color='red', linestyle='--', label=f'Most Common: {most_common_recovery_time} months')

# Step 8: Add labels and title
plt.title("Distribution of Recovery Times for Hand & Elbow Injuries")
plt.xlabel("Recovery Time (Months)")
plt.ylabel("Frequency")
plt.legend()

# Show the plot
plt.show()
```
